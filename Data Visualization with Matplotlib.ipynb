```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph
plt.figure(figsize=(5,3), dpi=150)


#plt.plot(x,y, label='2x', color='blue', linewidth=2, marker='.', linestyle='dashed' ,markersize=10, markeredgecolor='blue')
plt.plot(x,y, 'b^--', label='2x')

# Line number 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')

plt.title('First Graph')
plt.xlabel('X Axis', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.ylabel('Y Axis', fontdict={'fontname': 'ARIAL'})

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8])

plt.legend()

plt.savefig('mygraph.png', dpi=300)

plt.show()
```

## Bar Chart

```python
labels = ["A", "B", "C"]
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', '-', '*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))



plt.legend

plt.show()
```

## Real World Problems 

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.plot(gas.Year, gas.USA, 'r.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'b.-', label='Canada')

plt.title('USA vs Canada Gas Prices', fontdict={'fontweight': 'bold', 'fontsize': 18})
plt.ylabel('$/Gallon')
plt.xlabel('Year')
plt.legend()

plt.xticks(gas.Year[::3].tolist())

plt.show()

```

```python
plt.figure(figsize=(8,5))
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], label=country)
plt.title('Gas prices (USD)')
plt.legend()
plt.xticks(gas.Year[::3].tolist())
plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

## Histograms

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#ababab')
plt.xticks(bins)

plt.style.use('ggplot')
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')
plt.show()
```

## Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['left', 'right']
colors = ['#abcdef', '#eeefad']

plt.style.use('ggplot')
plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f %%')
plt.title('Foot Preference of Fifa Players')
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight > 200].count()[0]


weights = [light, light_medium,  medium, medium_heavy, heavy]
labels = ['< 125', '125-150', '150-175', '175-200', '>200']
explode = (.4,.1,0,0,.4)

plt.title("Weights Distribution of FIFA players (lbs)")
plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)
plt.legend()
plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(5,6))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['Fc Barcelona', 'Real Madrid', 'NE Revolution']

boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    # Set edge color 
    box.set(color='#4286f4', linewidth=2)
    
    # Change Fill color
    box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall rating')
plt.xlabel('Teams')

plt.show()
```
