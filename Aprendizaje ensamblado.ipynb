```python
from __future__ import division
import pandas as pd
import numpy as np

churn_df = pd.read_csv(r"C:\Users\cxb0254\Documents\Cursos_dat-academy\Pracice_Python\Churn.csv", sep = ';')
col_names = churn_df.columns.tolist()

print ("Column names:")
print (col_names)

churn_df.head(6)
```

```python
y = churn_df['Churn']
print("Churns positivos: ", sum(y))
print("Churns negativos: ", y.shape[0] - sum(y))

features = churn_df.columns
X = churn_df.as_matrix().astype(np.float)
print("Hay %d ejemplos y %d variables" %X.shape)
```

## RANDOM FOREST

```python
np.random.seed(10)
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
X_train, X_test = X[0:2500], X[2501:]
y_train, y_test = y[0:2500], y[2501:]
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

n_estimator = 1
rf = RandomForestClassifier(max_depth=30, n_estimators=n_estimator)
rf.fit(X_train, y_train)
yhat = rf.predict(X_test)
print("Report (1 Ã¡rbol): \n", metrics.classification_report(y_test,yhat))

n_estimator = 10
rf = RandomForestClassifier(max_depth=30, n_estimators=n_estimator)
rf.fit(X_train, y_train)
yhat = rf.predict(X_test)
print("Report (10 Ã¡rboles): \n", metrics.classification_report(y_test,yhat))

n_estimator = 100
rf = RandomForestClassifier(max_depth=30, n_estimators=n_estimator)
rf.fit(X_train, y_train)
yhat = rf.predict(X_test)
print("Report (100 Ã¡rboles): \n", metrics.classification_report(y_test,yhat))
```
