```
import pytesseract
import shutil
import os
import random
try:
 from PIL import Image
except ImportError:
 import Image
```

## Example 1

```
image_path_in_colab = '..//letras.jpg'
image = Image.open(image_path_in_colab)
image = image.convert('RGBA')
pixel = image.load()
```

```
image.size
```

```
image
```

```
image = image.resize((120, 60))
```

```
image
```

```
for y in range(image.size[1]):
  for x in range(image.size[0]):
    if pixel[x, y][0] < 102 or pixel[x, y][1] < 102 or pixel[x, y][2] < 102:
      pixel[x, y] = (0, 0, 0, 255)
```

```
image = image.convert('RGB')
image.save('g.jpg')
out = pytesseract.image_to_string(Image.open('g.jpg'))
```
